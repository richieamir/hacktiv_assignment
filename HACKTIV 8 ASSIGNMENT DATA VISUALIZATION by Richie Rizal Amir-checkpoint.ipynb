## Visualization and Analyze 

### A. Area plot

__Question :__ How many crime per category every month in 2015 at Greenwhich?

To answer this question, I must create dataframe consist only borough = Greenwhich, major category, value > 1 and year on 2015

#### Step A.1 Masking dataframe

```python
df_area_plot = df[(df.borough == 'Greenwich') 
                  & (df.value > 0) 
                  & (df.year == 2016)
                 ]
df_area_plot.head()
```

#### Step A.2 See info if there is null column or data type that doesn't fit to its column

```python
df_area_plot.info()
```

#### Step A.3 Remove unnecessary columns

```python
# deleting year because all the year have been 2015
df_area_plot.drop(['lsoa_code','minor_category','year','borough'],axis=1,inplace=True)
df_area_plot.head()
```

#### Step A.4 Set major category value to columns  and index to months

```python
new_columns = df_area_plot['major_category'].unique()
months = list(range(1,13))

df_area_plot_new = pd.DataFrame(columns=new_columns,index=months)

for col in df_area_plot_new.columns:
    for row in df_area_plot_new.index:
        val = df_area_plot[(df_area_plot['major_category'] == col) & (df_area_plot['month'] == row)]['value'].sum()
        df_area_plot_new.loc[row,col] = val 
df_area_plot_new.head()
```

#### Step A.5 Visualize Area Chart

```python
ax = df_area_plot_new.plot.area(figsize=(14,10))

plt.title('Number of Crimes in Greenwhich on 2015')
plt.ylabel('Number of Crimes')
plt.xlabel('Months')

plt.show()
```

```python
ax = df_area_plot_new.plot.area(stacked=False,figsize=(14,10))

plt.title('Number of Crimes in Greenwhich on 2015')
plt.ylabel('Number of Crimes')
plt.xlabel('Months')

plt.show()
```

__Insight__ : The most criminal has been commited is __'Violence Against the Person'__ and followed by __'Theft and Handling'__. Average total of crime per month is arround __1750 til 2000__. __'Theft and Handling'__ is incresing significantly in __May and October__. Violence Against the Person is decreasing in __September__. The least crime has been commited is __Robery__ and followed by __'Other Notifiable Offences'__

### C. Bar Chart

__Question :__ Which borough has the most and the least crime in 2015?

#### Step C.1 Prepare dataframe consist borough and value in 2015

```python
df_bar = df[(df['year'] == 2015) & (df['value'] > 0)]
df_bar = df_bar[['borough','value']]

new_bar_index = df_bar['borough'].unique()

new_bar_new = pd.DataFrame(columns=['value'],index=new_bar_index)

for row in new_bar_new.index:
    new_bar_new.loc[row]['value'] = df_bar[df_bar['borough'] == row]['value'].sum(axis=0)
```

##### C.2  Visualize

```python
new_bar_new.plot(kind='bar');
plt.title('Number of crimes every borough')
plt.ylabel('Crimes')
plt.show()
```

__Insight :__ From the chart above, I conclude *Westminster* has the most crime rate in London. The least crime rate in London i *Kingston upon Thames*. 

### D. Pie Chart

__Question :__ Which has the most crime category in Westminster ?

#### D.1 Create dataframe 

```python
#Masking data from Westminster
df_pie = df[(df['borough'] == 'Westminster') & (df['value'] > 0)]

pie_columns = df_pie['major_category'].unique()
df_pie_new = pd.DataFrame(columns=['value'],index=pie_columns)

#Group every Crime Category
for row in df_pie_new.index:
    result = df_pie[df_pie['major_category'] == row]['value'].sum(axis=0)
    df_pie_new.loc[row]['value'] = result
```

##### D.2 Visualize 

```python
plt.pie(df_pie_new['value'],labels=df_pie_new.index,autopct='%1.1f%%')
plt.title('Percentage every crime category in Westminster')
plt.show()
```

__Insight :__  Theft and Handling is the most crime has been commited in Wesminter with 61%. Other notifiable has the least crime has been commited followed by Robery with 3.5%.

### G. WordCloud

__Question :__ What is the most minor category that occur in dataframe?

#### G.1 Ctreate set contains stopwords

```python
stopwords = set(STOPWORDS)
```

#### G.2 Creating corpus

```python
corpus = df['minor_category'].values
corpus = ' '.join(corpus)
```

#### G.3 Generating wordcloud

```python
df_wc = WordCloud(background_color='White',
                     max_words=2000,
                     stopwords=stopwords)
df_wc.generate(corpus)
```

```python
plt.imshow(df_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

__Insight :__ The most minor category is Criminal Damage

__Overal Analisis :__ Crime rates in london is decreasing every year. It makes London has great improvement in security. The most crime has been commit is Criminal Damage and Violence Against the Person that consist *Offensive weapon, comman assault, harrasment, damage to motor vehicle, damage to building, and wounding*. London could improve their security in Westminster because it has the most crime has been committed.
